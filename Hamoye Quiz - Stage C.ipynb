#### Cleaning

```python
# Converting 'Year' column to object datatype
df['Year'] = df['Year'].astype(object)
```

```python
# grouping datasets by Country column
grp = df.groupby('Country')
grp.first()
```

```python
# grouping datasets by Country & Year columns

grp2 = df.groupby(['Country', 'Year'])
grp2.first().head(60)
```

```python
# creating Sum of Food Items per Year as Value_sum column 
# grouping and transforming Value column

df['Value_sum'] = df.groupby(['Country', 'Year']).Value.transform('sum')
df.head()
```

```python
# copying datasets
viz = df.copy(['Country', 'Year', 'Value_sum'])
viz
```

```python
# dropping redundant columns
viz.drop(['Item', 'Value'], axis = 1, inplace = True)
```

```python
# confirming drop
viz.head()
```

```python
# grouping dataset by Country & Year columns

gviz = viz.groupby(['Country', 'Year'])
gviz.first()
```

```python
# visualizing a distribution of Value_sum column

# Plot size
plt.figure(figsize=(15,8))

# Plotting chart
plt.hist(data=viz, x='Value_sum', density=True, bins=35)

# Title & Axes labels
plt.title('Distribution of Value_sum Column', fontsize=15)
plt.ylabel('Count', fontsize=13)
plt.xlabel('Value_sum', fontsize=13);
```

```python
# visualizing outliers using boxplots

# Plot size
plt.figure(figsize=(15,8))

# Plotting chart
plt.boxplot(data=viz, x='Value_sum')

# Title & Axes labels
plt.title('Distribution of Value_sum Column', fontsize=15, weight=2)
plt.ylabel('Count', fontsize=13)
plt.xlabel('Value_sum', fontsize=13)

# Limits for the Y axis
plt.ylim(0,80000);
```

#### Insights

- We can see there are numerous outliers in this dataset. 
- The mean of our data is within 20Kt
- The upper fence is within 170Kt - hence, values beyond the upper fence are considered to be outliers.
- Outliers need will be removed to ensure high quality of data for analysis

### Average and median food production, the outliers, Quartile and interquartile ranges. (preferably a Boxplot).

```python
# setting plot color parameter to a default
plot_color = sns.color_palette()[0]

# Plotting chart
plt.figure(figsize=(14.70,8.27))
ax = sns.boxplot(data = viz, x = 'Year', y = 'Value_sum', 
           color = plot_color);

#renaming axes
ax.set(xlabel = 'Listing Year', ylabel = 'BorrowerRate');

#renaming axes and titles
ax.set_title("Trend Showing Annual Distribution of Food Production",fontsize=15, weight ='bold')
ax.set_xlabel('Year', fontsize=13, weight ='bold');
ax.set_ylabel('Food Production', fontsize=13, weight ='bold');

# Limits for the Y axis
plt.ylim(0,70000);
```
